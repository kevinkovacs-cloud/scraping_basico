```python
page= requests.get("https://parascrapear.com/")
soup=BeautifulSoup(page.text, "html.parser") #defino la sopa, atributo .text y defino el parser para interpretar la estructura
#una vez que tengo la sopa voy a la busqueda
blockquote_items=soup.find_all("blockquote") #utilizamos otra variable, en este caso blockquote_items
#dentro de sopa, todos los atributos "blockquote" con el metodo find_all
#en blockquote estaban todos los datos a scrapear

for blockquote in blockquote_items:
  autor = blockquote.find (class_="author").text
  print([autor])
```

```python
for blockquote in blockquote_items:
  autor = blockquote.find (class_="author").text
  categoria= blockquote.find(class_="cat").text
  print([autor, categoria])
```

```python
for blockquote in blockquote_items:
  autor = blockquote.find (class_="author").text
  categoria= blockquote.find(class_="cat").text
  frase = blockquote.find("q").text
  print([autor, categoria, frase])
```

x con c/u de los elementos scrapeados de https://parascrapear.com
 hacemos listas

```python
lista_autor=[]
lista_categoria=[]
lista_frase=[]
for blockquote in blockquote_items:
  autor = blockquote.find (class_="author").text
  lista_autor.append(autor)
  categoria= blockquote.find(class_="cat").text
  lista_categoria.append(categoria)
  frase = blockquote.find("q").text
  lista_frase.append(frase)
print(lista_autor)
print(lista_categoria)
print(lista_frase)
```

Creamos el DataFrame:

```python
datos = {'autor': lista_autor, 'categoria': lista_categoria, 'frase': lista_frase}
df = pd.DataFrame(datos)
df
```

Ordenamos:

```python
df_sorted = df.sort_values('autor')
df_sorted
```
